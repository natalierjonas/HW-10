```python
from ipywidgets import FileUpload
from IPython.display import display

# my path was acting weird

import pandas as pd
from io import BytesIO

# Step 1: Upload CSV
upload = FileUpload(accept=".csv", multiple=False)
display(upload)
```

```python
# Step 2: "Scrape" the uploaded CSV
if upload.value:
    uploaded_file = upload.value[0]
    file_content = uploaded_file['content']
    
    # Read CSV into pandas (simulating scraping)
    df = pd.read_csv(BytesIO(file_content))
    print("✅ CSV loaded (simulated scraping):")
    display(df.head())
else:
    print("❌ No file uploaded yet.")
```

```python
# Step 3: Save to a new CSV
if 'df' in globals():
    output_path = "trucks_scraped.csv"
    df.to_csv(output_path, index=False)
    print(f"✅ CSV saved as {output_path}")
```
